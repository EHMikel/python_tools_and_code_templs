# MLP with Keras for REGRESSION TASK
## Building an Housing predicting value Using the Sequential API

```python
import numpy as np

import sys

import matplotlib.pyplot as plt
import pandas as pd
import tensorflow.keras
import tensorflow as tf

print(f"Tensor Flow Version: {tf.__version__}")
print("GPU is", "available" if tf.test.is_gpu_available() else "NOT AVAILABLE")
#!nvidia-smi

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
```

### Load the data

```python
housing = fetch_california_housing()
print('PARTICION EXTERNA')
X_train_full, X_test, y_train_full, y_test = train_test_split(
                                        housing.data, housing.target,
                                        random_state=42, 
                                        test_size=0.25)
print('TRAIN DATA: ')
print(X_train_full.shape)
print(y_train_full.shape)
print('TEST DATA')
print(X_test.shape)
print(y_test.shape)
print('\n')
print('PARTICION INTERNA')

X_train, X_valid, y_train, y_valid = train_test_split(
                                X_train_full, y_train_full, 
                                random_state=42, 
                                test_size=0.25)
# PARTICION INTERNA
X_train, y_train = X_train_full[:-5000], y_train_full[:-5000]
X_valid, y_valid = X_train_full[-5000:], y_train_full[-5000:]
print('TRAIN DATA')
print(X_train.shape)
print(y_train.shape)
print('VALIDATION DATA')
print(X_valid.shape)
print(y_valid.shape)
```

### create, compile and fit - train the model

```python
from tensorflow.keras.models import Sequential                      # modelo que voy a usar
from tensorflow.keras.layers import Flatten, Dense, Normalization   # tipos de capas que voy usar
from tensorflow.keras.optimizers import Adam                        # optimizador que voy a usar

# create
norm_layer = Normalization(input_shape= X_train.shape[1:])
model = Sequential([norm_layer,                                     # input layer
                    Dense(40, activation= 'relu'),                  # HL 1
                    Dense(40, activation= 'relu'),                  # HL 2
                    Dense(40, activation= 'relu'),                  # HL 3
                    Dense(1)])                                      # output layer

#compile
optimizer = Adam(learning_rate= 0.001)
model.compile(optimizer= optimizer,
              loss = 'mse', 
              metrics= ['RootMeanSquaredError'])

# fit - train
norm_layer.adapt(X_train)
H = model.fit(X_train, y_train, 
              validation_data= (X_valid, y_valid), 
              epochs= 20, 
              batch_size= 16)
```

### Learning curves

```python
pd.DataFrame(H.history).plot(figsize= (10, 6), 
                            xlim= [0, 19], ylim= [0, 1], 
                            grid= True, 
                            xlabel= 'Epochs', 
                            style= ['r--', 'r','b--', 'b-^'])
plt.show()
```

### Evaluating the model

```python
from sklearn.metrics import r2_score
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
r2_score(y_test, y_pred)
```

## Using FUNCTIONAL API to build the model

Con la API funcional se le indica a cada capa, la capa anterior (es decir, a cada capa se le indica a cual esta conectada), pasandole el nombre de la capa anterior como un **argumento de una funcin**

### MODELO1 ---> MODO DUMMY

Primero se crearan las capas del modelo y luego se le indicarn a cual esta conectada cada capa, pero es una practica habitual hacerlo a la vez.

Se unen las capas de manera secuencial y adems se una la capa de normalizacin con la capa de salida, es decir, hay dos caminos el superficial y el profundo, pero los mismos datos pasar por ambos caminos. Este caso se entrenan a la vez un perceptron simple (con una sola neurona: la de salida) y uno multicapa.

#### Create, compile, and fit - train the model

```python
from tensorflow.keras.layers import Normalization, Dense, Concatenate, Input
from tensorflow.keras import Model
tf.random.set_seed(42)
# create
norm_layer = Normalization()                            # to standarize the inputs
hl_1 = Dense(30, activation= 'relu')                    # hidden layer 1
hl_2 = Dense(30, activation= 'relu')                    # hidden layer 2
concat_layer= Concatenate()                             # para concatenar
output_layer = Dense(1)                                 # output layer

# construct
input_ = Input(shape=X_train.shape[1:])                 # se le indica el shape y el dtype
normalized = norm_layer(input_)                         # a la norm layer se la pasa la input
hidden_1 = hl_1(normalized)                             # la hl 1 se conecta a la norm layer
hidden_2 = hl_2(hidden_1)                               # la hl 2 se conecta a la hl1
concat = concat_layer([normalized, hidden_2])           # concatenamos la norm layer con la hl2
output = output_layer(concat)                           # el output se conecta a la concat layer

model = Model(inputs= [input_], outputs= [output])      # construimos el modelo desde el input hasta el outpul

# compile
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer = optimizer, 
              loss= 'mse', 
              metrics= ['RootMeanSquaredError'])

#fit
norm_layer.adapt(X_train)
H1 = model.fit(X_train, y_train,
               validation_data= (X_valid, y_valid), 
               epochs= 20, 
               batch_size= 32)
```

### Learing curves

```python
pd.DataFrame(H.history).plot(figsize= (10, 6), 
                            xlim= [0, 19], ylim= [0, 1], 
                            grid= True, 
                            xlabel= 'Epochs', 
                            style= ['r--', 'r','b--', 'b-^'])
plt.show()
```

#### Evaluate the model

```python
model.evaluate(X_test, y_test)
```

```python
y_pred_m1 = model.predict(X_test)

print(r2_score(y_test, y_pred_m1)) # la r2 score a bajado, creo que es debido a que hemos usado 30 neuronas en vez de 40 
```

### MODELO2 ---> Creando y Construyendo

Ahora se creara y construira el modelo a la vez. En las capas de entrada no se puede hacer por que despues le tenemos qeu meter el mtodo adapt para normalizar los datos.

Esta vez tambien habr dos caminos el superficial y el profundo, pero no pasarn los mismos datos por el mismo camino. Este caso se entrenan a la vez un perceptron simple (con una sola neurona: la de salida) para ciertas caractersticas y uno multicapa para otras caractersticas.

#### Create, compile, and fit - train the model

```python
print(X_train.shape)
```

```python
X_train_wide, X_train_deep = X_train[:, :5], X_train[:, 2:]     # cojemos las 5 primeras features para el camino corto
X_valid_wide, X_valid_deep = X_valid[:, :5], X_valid[:, 2:]     # las 6 ltimas features para el camino largo (deep)
X_test_wide, X_test_deep = X_test[:, :5], X_test[:, 2:]

print( X_train_wide.shape, X_train_deep.shape)
print( X_valid_wide.shape, X_valid_deep.shape)
print( X_test_wide.shape, X_test_deep.shape)
```

```python
from tensorflow.keras.layers import Normalization, Dense, Input, concatenate
from tensorflow.keras import Model
from tensorflow.keras.optimizers import Adam

tf.random.set_seed(42)
# create - construct
input_wide= Input(shape=[5])
input_deep= Input(shape=[6])
norm_layer_wide= Normalization()
norm_layer_deep= Normalization()
norm_wide = norm_layer_wide(input_wide)
norm_deep= norm_layer_deep(input_deep)
hidden_layer_1 = Dense(30, activation= 'relu')(norm_deep)
hidden_layer_2 = Dense(30, activation= 'relu')(hidden_layer_1)
concat = concatenate([norm_wide, hidden_layer_2])
output = Dense(1)(concat)
functional_model = Model(inputs= [input_wide, input_deep], outputs= output)

# compile 
optimizer = Adam(learning_rate=0.001)
functional_model.compile(optimizer= optimizer, 
                         loss= 'mse', 
                         metrics= ['RootMeanSquaredError'])

# fit - train
norm_layer_wide.adapt(X_train_wide)
norm_layer_deep.adapt(X_train_deep)
H_deep_wide = functional_model.fit((X_train_wide, X_train_deep), y_train,
                                    validation_data= ((X_valid_wide, X_valid_deep), y_valid), 
                                    epochs= 20, 
                                    batch_size= 32)
```

### learning curves

```python
pd.DataFrame(H_deep_wide.history).plot(figsize= (10, 6), 
                            xlim= [0, 19], ylim= [0, 1], 
                            grid= True, 
                            xlabel= 'Epochs', 
                            style= ['r--', 'r','b--', 'b-^'])
plt.show()
```

### evaluate the model

```python
y_pred_wide_deep = functional_model.predict((X_test_wide, X_test_deep))

r2_score(y_test, y_pred_wide_deep)
```
